```python
import pandas as pd
```

```python
data =  pd.read_csv('Data/realtor-data.csv')
data.head()
```

```python
data.dropna(inplace=True)
data.drop('prev_sold_date', axis=1, inplace=True)
data.drop('brokered_by', axis=1, inplace=True)
data.drop('status', axis=1, inplace=True)
data.drop('zip_code', axis=1, inplace=True)
data.drop('street', axis=1, inplace=True)
```

```python
city_codes = pd.DataFrame(data['city'].astype('category').cat.codes)
city_names = pd.DataFrame(data['city'])
city_state = pd.DataFrame(data['state']).join(city_names).set_index('state')
data = data.join(pd.get_dummies(data.state)).drop(['state'], axis=1)
data['city'] = data['city'].astype('category').cat.codes
```

```python
city_state = city_state.drop_duplicates()
```

```python
city_names_to_codes = city_names.join(city_codes)
city_names_to_codes.rename(columns = {'city':'city', 0:'city code'}, inplace=True)
city_names_to_codes = city_names_to_codes.set_index('city').drop_duplicates()
city_state = city_state.reset_index().set_index('city')
```

```python
city_names_to_codes.to_csv('Data/city_codes.csv')
city_names_to_codes
```

```python
city_state_data = city_state.join(city_names_to_codes).reset_index().set_index('city code')
city_state_data.to_csv('Data/city-state-data.csv')
city_state_data
```
